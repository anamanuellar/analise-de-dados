**1. Você percebeu que seria melhor alterar o nome de algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.**



```python
df.rename(columns={'ID': 'Identificador', 'Pressão sanguíneaaaa': 'Pressão Sanguínea',
                   'Ocupação': 'Profissão', 'Categoria BMI': 'Categoria IMC'}, inplace=True)
df.head()
```

**2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]**



```python
#Opção de resposta 1
resumo_horas_sono = df.groupby('Profissão')['Duração do sono'].agg(['mean', np.median, pd.Series.mode])
print("Resumo de Horas de Sono para cada Profissão:")
resumo_horas_sono
```

```python
#Opção de resposta 2
medias_sono = df.groupby('Profissão')['Duração do sono'].mean()
modas_sono = df.groupby('Profissão')['Duração do sono'].agg(lambda x: pd.Series.mode(x).tolist())
medianas_sono = df.groupby('Profissão')['Duração do sono'].median()
medias_sono
```

**3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?**

```python
df['Obeso'] = df['Categoria IMC'].isin(['Obesidade'])
porcentagem_obesos_eng_software = df[df['Profissão'] == 'Eng. de Software']['Obeso'].mean() * 100
print("Porcentagem de Obesos entre os Engenheiros de Software:")
print(f"{porcentagem_obesos_eng_software:.2f}%")
```

**4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)**

```python
media_horas_sono_por_profissao = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])].groupby('Profissão')['Duração do sono'].mean()
media_horas_sono_por_profissao
```

```python
resposta4 = "Advogados dormem menos." if media_horas_sono_por_profissao['Advogado(a)'] < media_horas_sono_por_profissao['Representante de Vendas'] else "Representantes de vendas dormem menos."
resposta4
```

**5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)**

```python
media_horas_sono_por_profissao = df[df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])].groupby('Profissão')['Duração do sono'].mean()
media_horas_sono_por_profissao
```

```python
resposta5 = "Quem fez enfermagem tem menos horas de sono." if media_horas_sono_por_profissao['Enfermeiro(a)'] < media_horas_sono_por_profissao['Médico(a)'] else "Quem fez Medicina tem menos horas de sono."
resposta5
```

**7. Descubra qual a profissão menos frequente no conjunto. (Use

**8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)**

```python
df['Pressão Sanguínea'] = df['Pressão Sanguínea'].apply(lambda x: int(x.split('/')[0]) if isinstance(x, str) else x)
media_pressao_sanguinea_genero = df.groupby('Gênero')['Pressão Sanguínea'].mean()
media_pressao_sanguinea_genero
```

```python
resposta8 = "As mulheres têm maior pressão sanguínea." if media_pressao_sanguinea_genero['Mulher'] > media_pressao_sanguinea_genero['Homem'] else "Os homens têm maior pressão sanguínea."
resposta8
```

**9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?**

```python
moda_horas_sono = df['Duração do sono'].mode().iloc[0]
moda_horas_sono
```

```python
resposta9 = "É predominante entre os participantes dormir 8 horas por dia." if moda_horas_sono == 8 else "Não é predominante entre os participantes dormir 8 horas por dia."
resposta9
```

**10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)**

```python
media_passos_frequencia_cardiaca = df.groupby(df['Frequência cardíaca'] > 70)['Passos diários'].mean()
media_passos_frequencia_cardiaca
```

```python
resposta10 = "Pessoas com frequência cardíaca menor ou igual a 70 dão mais passos em média." if media_passos_frequencia_cardiaca[False] > media_passos_frequencia_cardiaca[True] else "Pessoas com frequência cardíaca acima de 70 dão mais passos em média."
resposta10
```
